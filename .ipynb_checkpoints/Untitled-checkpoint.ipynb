```python
#This file will examine charts of the top 200 songs globally and from 62 countries for the week of 07/10/2020
#to 07/16/20 to determine if there is a significant relation between where a song becomes popular and its global
#relevance.


# Dependencies
import pandas as pd
import os
import csv
import matplotlib.pyplot as plt
import scipy.stats as sts
import numpy as np
import sys
import spotipy
import spotipy.util as util
from spotipy.oauth2 import SpotifyClientCredentials
import json
from scipy.stats import linregress
import numpy as np
from sklearn import datasets
```

```python
#Now working on making the data more readable in the files so that we can look at it holistically
chart_list = ['Australia','Sweden','Canada','UnitedKindom','Germany','India','France','Ireland','Italy']
#looping to save time yay

    
countries_df = {'country' + str(i):pd.read_csv('GlobalCharts/Top50_' + chart_list[i-1] + '_7-9-2020.csv') for i in range(1,10)}

total_chart_df = pd.read_csv('GlobalCharts/Top50_USA_7-9-2020.csv')
```

```python
#Now let's repeat the process we did with the global chart with all the files

for country in chart_list:
    
    country_df = pd.read_csv('GlobalCharts/Top50_' + country + '_7-9-2020.csv')
    country_df = country_df.drop(columns=['Track Name', 'Artist'])
    country_df = country_df.rename(columns = {'Position' : country + ' - Rank', 'URL' : country + ' URL'})
 
    #total_chart_df = pd.concat([total_chart_df, country_df], axis=1)
    total_chart_df = pd.merge(total_chart_df, country_df, how='left', left_on='URL', right_on=country + ' URL') #this merges on song name
    total_chart_df.drop(columns=[country + ' URL'],inplace=True)

#final chart
total_chart_df=total_chart_df.fillna(0)
#total_chart_df_t = total_chart_df.transpose()
#total_chart_df_t[0]
```

```python
#Adding function to plot information in a scatter plot
# inputs
#      xvalues = this is the x value series to be plotted
#      yvalues = This is the y value series to be plotted
#      x_loc = This is where we want the slope line to be displayed on the x axis
#      y_loc = This is where we want the slope line to be displayed on the y axis
#      x_label = Label for the x axis
#      y_label = Label for the y axis
#      title = Label for the Title of the plot

def plot_func(xvalues,yvalues,x_loc,y_loc,x_label,y_label,title,imagename='none'):
    
    # Generate the pearson coeeficient
    correlation = sts.pearsonr(xvalues,yvalues)
    
    # display the information to the screen
    print(f"The correlation between both factors is {round(correlation[0],2)}")
    print(f"The p-value is {round(correlation[1],3)}")

    # Get the linear regression model
    (slope, intercept, rvalue, pvalue, stderr) = linregress(xvalues, yvalues)

    # Create the line using the slope and intercept and print out the formula
    regress_values = xvalues * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

    # Plot information
    fig = plt.figure()
    plt.scatter(xvalues,yvalues)
    plt.plot(xvalues,regress_values,"r-")
    plt.annotate(line_eq,(x_loc,y_loc),fontsize=15,color="red")
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.show()
    
    # Save the image
    if imagename!= 'none':
        fig.savefig(imagename)
```
